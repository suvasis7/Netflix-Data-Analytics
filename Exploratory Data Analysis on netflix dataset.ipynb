<H2>Step 2: Data Preparation Step</H2>

```python
miss=data.isnull().sum()
```

```python
miss
```

<h3>Observation:</h3>

<i>This tells us how many data is missing in each of the 12 columns we have.</i>

```python
missing=(data.isnull().sum()/len(data))*100
```

```python
missing
```

```python
missing1=pd.concat([miss,missing],axis=1,keys=['Total_missing','Missing%'])
```

```python
missing1
```

<h3>Observation:</h3>

<i>The 'Total_missing' column tells us how many rows have missing data in each column and the 'missing%' column 
tells the percentage of the missing data in each column like director column is missing 2634 missing rows which amounts to 29.90% of the total data and so for other columns like cast and country.</i>

<h3>Heatmap for checking missing values</h3>

```python
plt.figure(figsize=(6,5),dpi=200)
sns.heatmap(data.isnull(),cmap='viridis')
```

<h3>Observation:</h3>

<i>From the above heatmap we can infer that director, cast and country columns are the with 
maximum missing values.</i>
